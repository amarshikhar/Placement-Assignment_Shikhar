```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Step 2: Load the dataset
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTwrFfv5h0Jq1SwWuLK83Fl-hUn-j2c40aGXMwgGHofbGivITrekXX-EGz0gNv0NdT26YrW26a-8NmN/pub?gid=1728606945&single=true&output=csv')  # Replace 'your_dataset.csv' with the actual file path

# Step 3: Perform data preprocessing
# Handle missing values if any
df = df.dropna()

# Convert categorical variables to numerical representations if needed
# (e.g., using one-hot encoding or label encoding)
df.head()
```

```python
df.columns=['Unnamed: 0', 'S.No', 'username', 'caption', 'hashtag', 'followers',
       'time_since_posted', 'likes']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
# Feature engineering - Extract features from text data (caption and hashtag)
# Combine caption and hashtag into a single feature
df['text_features'] = df['caption']

# Split the dataset into input features and target variables
X_text = df['text_features']
X_numeric = df[['followers']]
y_likes = df['likes']
y_time_since_posted = df['time_since_posted']

# Step 4: Split the dataset into training and testing sets
X_text_train, X_text_test, X_numeric_train, X_numeric_test, y_likes_train, y_likes_test, y_time_train, y_time_test = train_test_split(
    X_text, X_numeric, y_likes, y_time_since_posted, test_size=0.2, random_state=42)

# Step 5: Apply NLP techniques and feature engineering
# Apply text vectorization using CountVectorizer
vectorizer = CountVectorizer()
X_text_train_vectorized = vectorizer.fit_transform(X_text_train)
X_text_test_vectorized = vectorizer.transform(X_text_test)

# Combine vectorized text features with numeric features
X_train = pd.DataFrame(X_text_train_vectorized.toarray(), columns=vectorizer.get_feature_names_out()).join(X_numeric_train.reset_index(drop=True))
X_test = pd.DataFrame(X_text_test_vectorized.toarray(), columns=vectorizer.get_feature_names_out()).join(X_numeric_test.reset_index(drop=True))

# Step 6: Select a model architecture
model_likes = LinearRegression()
model_time_since_posted = LinearRegression()

# Step 7: Train the model
model_likes.fit(X_train, y_likes_train)
model_time_since_posted.fit(X_train, y_time_train)

# Step 8: Evaluate the model
likes_predictions = model_likes.predict(X_test)
time_since_posted_predictions = model_time_since_posted.predict(X_test)

mse_likes = mean_squared_error(y_likes_test, likes_predictions)
mae_time_since_posted = mean_absolute_error(y_time_test, time_since_posted_predictions)

print(f"Mean Squared Error (Likes): {mse_likes}")
print(f"Mean Absolute Error (Time Since Posted): {mae_time_since_posted}")

# Step 9: Fine-tune the model (optional)
# Experiment with hyperparameter tuning or try different models to improve performance

# Step 10: Make predictions on new data
new_text_data = ['new_caption new_hashtag']
new_numeric_data = pd.DataFrame({'followers': [1000]})
new_text_data_vectorized = vectorizer.transform(new_text_data)

new_data = pd.DataFrame(new_text_data_vectorized.toarray(), columns=vectorizer.get_feature_names()).join(new_numeric_data)
likes_prediction_new_data = model_likes.predict(new_data)
time_since_posted_prediction_new_data = model_time_since_posted.predict(new_data)

print(f"Likes Prediction for New Data: {likes_prediction_new_data}")
print(f"Time Since Posted Prediction for New Data: {time_since_posted_prediction_new_data}")
```
